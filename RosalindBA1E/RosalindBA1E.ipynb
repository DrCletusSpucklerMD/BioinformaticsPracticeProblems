```python
def patternCount(text, pattern):
    count = text.count(pattern)
    return count
```

```python
def find_clump(genome, k, L, t):
    k_mers = []
    for i in range(len(genome) - L + 2):
        window = genome[i:i+L]
        for j in range(len(window)):
            k_mer = window[j:j+k]
            if len(k_mer) == k and patternCount(window, k_mer) == t:
                k_mers.append(k_mer)
    k_mers_unique = pd.unique(pd.Series(k_mers))
    return k_mers_unique
```
