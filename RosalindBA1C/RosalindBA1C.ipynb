```python
def reverse_complement(forward):
    reverse = []
    for i in range(len(forward)):
        if forward[i] == 'A':
            reverse.append('T')
        elif forward[i] == 'T':
            reverse.append('A')
        elif forward[i] == 'C':
            reverse.append('G')
        elif forward[i] == 'G':
            reverse.append('C')
        else:
            return 'error'
    reverse.reverse()
    reverse_string = ''.join(reverse)
    return reverse_string
```
